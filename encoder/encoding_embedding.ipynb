Encoding - Numerical representation (often sparse) of the data not necessarily preserving semantic meaning. --> Frquency based method.
Embedding - Meaningful numerical representation of the data. --> neural network/transformer based method.

Common encoding methods:
1. One-hot encoding
2. Bag of Words (BOW)
3. TF-IDF (Term Frequency & Inverse Document frequency)
4. N-grams
5. BM25: Improved version of TF-IDF, but it's ranking algo.
6. GloVe (Global Vectors) - Is a count based + matrix factorization-based

Vector - [1,2,3] --> Collection of values. Numerical representation of data in n-dimensional space.

In NLP, embedding is a Vector with meaning.


Similarity search --> 1. Euclidian distance between the points.
                (OR)  2. Dot product of the vectors [(x1*x2) + (y1*y2)]
                (OR)  3. cosine angle between the points




Classical word Embedding Models:
1. Word2Vec
2. FastText
The above 2 methods are based on Neural Networks in the early years (2018-2020) which were able to convert only word into embeddings (only text). They produce static embeddings (fixed vectors). They do not understand context.

Now, we have State of Art Techniques (SOTA) based on transformers which can convert any kind of data into embeddings like images, audio, video.


Example:
Sen 1: I sat on the river bank.
Sen 2: I deposited money in the bank.

In static embedding model (word2vec), the word "bank" is represented by the same vector in both the sentences. 
In SOTA transformer-based embeddings model, the word "bank" will be represented by a different vector becuase it understands the context of the sentences.


Feature                     Encoding                                            Embedding
-----------------------------------------------------------------------------------------------------------------

Nature                  Often sparse(count based)               Dense (Neural network or transformer based)
Semantic Meaning        Not preserved                           Preserved
Context Awareness       Not preserved                           Preserved
